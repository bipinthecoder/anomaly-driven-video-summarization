```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import cv2
import tensorflow as tf
```

```python
path = './Normal videos/Normal_Videos_345_x264.mp4'
```

```python
video_file = cv2.VideoCapture(path)
print(video_file)
```

```python
if not video_file.isOpened():
    print('Error reading video')
```

```python
frames_list = []
```

```python
while(video_file.isOpened()):
    #Reading frame-by-frame
    ret, frame = video_file.read()
    
    if not ret:
        print('Error in reading the frame or stream end')
        break
    else:
        frames_list.append(frame)
        
    if cv2.waitKey(1) == ord('q'):
        break
```

```python
video_file.release()
cv2.destroyAllWindows()
```

```python
print(len(frames_list))
```

```python
frame_to_display = frames_list[100]
plt.imshow(cv2.cvtColor(frame_to_display, cv2.COLOR_BGR2RGB))
plt.show()
```
